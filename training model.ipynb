```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read in the dataset
df = pd.read_csv('sent_train.csv')  # Adjust the file path as needed

# Mapping numeric labels to sentiment names
label_mapping = {0: 'Bearish', 1: 'Bullish', 2: 'Neutral'}
df['sentiment'] = df['label'].map(label_mapping)

# Set seaborn style
sns.set(style="whitegrid")

# Plotting the sentiment distribution
plt.figure(figsize=(7, 5))
sns.countplot(x="sentiment", data=df, palette="coolwarm", order=['Bearish', 'Bullish', 'Neutral'])
plt.title("Distribution of Sentiment Labels")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("sentiment_distribution.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import string
from collections import Counter

# Read in the dataset
df = pd.read_csv('sent_train.csv')

# Combine all text entries into one large string
all_text = ' '.join(df['text'].astype(str))

# Convert to lowercase and remove punctuation
all_text_clean = all_text.lower().translate(str.maketrans('', '', string.punctuation))

# Tokenize into words
words = all_text_clean.split()

# Count word frequencies
word_counts = Counter(words)

# Get the 20 most common words
top_50_words = word_counts.most_common(50)

# Convert to DataFrame for display
top_50_df = pd.DataFrame(top_50_words, columns=['Word', 'Frequency'])
print(top_50_df)
```

```python
import pandas as pd
import string
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Read in the dataset
df = pd.read_csv('sent_train.csv')

# Combine all text entries into one large string
all_text = ' '.join(df['text'].astype(str))

# Convert to lowercase and remove punctuation
all_text_clean = all_text.lower().translate(str.maketrans('', '', string.punctuation))

# Tokenize into words
words = all_text_clean.split()

# Count word frequencies
word_counts = Counter(words)

# Get the 20 most common words
top_40_words = word_counts.most_common(40)

# Convert to DataFrame
top_40_df = pd.DataFrame(top_40_words, columns=['Word', 'Frequency'])

# Set visual style
sns.set(style="whitegrid")

# Create barplot
plt.figure(figsize=(10, 6))
sns.barplot(x="Frequency", y="Word", data=top_40_df, palette="magma")
plt.title("Top 40 Most Frequent Words")
plt.xlabel("Frequency")
plt.ylabel("Word")
plt.tight_layout()

# Save the plot
plt.savefig("top_40_words.png", dpi=300)

# Show the plot
plt.show()
```
